```python
mode = "KEGG" # 选择使用kegg或者go进行富集分析

# 根据模式选择保存路径
kegg_parent_dir = "./kegg/"
go_parent_dir = "./go/"
if mode == "kegg" or mode == "KEGG":
    print("执行KEGG富集分析")
    parent_dir = kegg_parent_dir
else:
    print("执行GO富集分析")
    parent_dir = go_parent_dir
img_dir = "%s/img"%parent_dir
dataframe_dir = "%s/dataframe"%parent_dir # 后续读取这里的文件
```

```python
import pandas as pd

keyword = "Cord" # 关键词输入

k = 8
filtered_df_dict = {}
key_filtered_df_dict = {}
for i in range(k):
    title = "Clusters%d"%i
    print(title)
    df_filename = "%s/%s.csv"%(dataframe_dir, title)
    df = pd.read_csv(df_filename, index_col=0)
    filtered_df = df[df["Adjusted P-value"]<=0.05]
    key_filtered_df = filtered_df[filtered_df["Term"].apply(lambda x: keyword.lower() in x.lower())]
    filtered_df_dict[i] = filtered_df
    key_filtered_df_dict[i] = key_filtered_df
```
