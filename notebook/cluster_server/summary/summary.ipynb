```python
import pandas as pd
```

```python
erythroid_lineage_df = pd.read_csv("./metric/erythroid_lineage.csv")
gastrulation_df = pd.read_csv("./metric/gastrulation.csv")
```

```python
gastrulation_df["Model"]+"_gastrulation"
```

```python
erythroid_lineage_df["Dataset"] = "erythroid_lineage"
gastrulation_df["Dataset"] = "gastrulation"
```

```python
df = pd.concat([erythroid_lineage_df, gastrulation_df])
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(2, 1, figsize=(15,10))
sns.boxplot(
    df.loc[df["Metric"].isin(["CBDir"])],
    orient="v", x="Dataset", y="Score", hue="Model", 
    ax=ax[0], showmeans=True,
    meanprops={"marker":"o",
                "markerfacecolor":"white", 
                "markeredgecolor":"black",
                "markersize":"8"},)
ax[0].set_title("CBDir")
sns.move_legend(ax[0], "upper left", bbox_to_anchor=(1.0, 0)) # 移动图例(1,1)表示图形右上端

sns.boxplot(
    df.loc[df["Metric"].isin(["ICVCoh"])],
    orient="v", x="Dataset", y="Score", hue="Model", 
    ax=ax[1], showmeans=True,
    meanprops={"marker":"o",
                "markerfacecolor":"white", 
                "markeredgecolor":"black",
                "markersize":"8"})
ax[1].set_title("ICVCoh")
ax[1].legend_.remove()
```
