```python
mode = "GO" # 选择使用kegg或者go进行富集分析

# 根据模式选择保存路径
kegg_parent_dir = "./kegg/"
go_parent_dir = "./go/"
if mode == "kegg" or mode == "KEGG":
    print("执行KEGG富集分析")
    parent_dir = kegg_parent_dir
else:
    print("执行GO富集分析")
    parent_dir = go_parent_dir
img_dir = "%s/img"%parent_dir
dataframe_dir = "%s/dataframe"%parent_dir # 后续读取这里的文件
```

```python
import os
import pandas as pd

keyword = "Blood" # 关键词输入

if mode == "KEGG":
    k =  len(os.listdir(img_dir))
else:
    k =  int(len(os.listdir(img_dir))/4)
    
filtered_df_dict = {}
key_filtered_df_dict = {}
for i in range(k):
    title = "Clusters%d"%i
    print(title)
    df_filename = "%s/%s.csv"%(dataframe_dir, title)
    df = pd.read_csv(df_filename, index_col=0)
    filtered_df = df[df["Adjusted P-value"]<=0.05]
    key_filtered_df = filtered_df[filtered_df["Term"].apply(lambda x: keyword.lower() in x.lower())]
    filtered_df_dict[i] = filtered_df
    key_filtered_df_dict[i] = key_filtered_df
```

## 这里基因聚类内部显著的top的通路作为参考


```python
top = 3
pathway_list = []

for k in filtered_df_dict.keys():
    df = filtered_df_dict[k].sort_values("Adjusted P-value") # 根据P值排序
    pathway_list += list(df["Term"][:top])
    # pathway_list += list(filtered_df_dict[k].loc[:top]["Term"])

pathway_list = list({}.fromkeys(pathway_list).keys()) # 使用字典进行列表去重
pathway_list
```
